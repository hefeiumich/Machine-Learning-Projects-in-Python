# Select multiple columns from table

```python
import sqlite3
import pandas as pd
con = sqlite3.connect("sakila.db")
```

```python
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df

query = ''' SELECT first_name, last_name FROM customer;'''

sql_to_df(query).head()
```

# Select all the columns from table with *

```python
query = ''' SELECT * FROM customer;'''
sql_to_df(query).head()
```

# Syntax for the SQL DISTINCT Statement

```python
query = ''' SELECT DISTINCT(country_id) FROM city'''
sql_to_df(query).head()
```

# Syntax for the SQL WHERE

```python
query = ''' SELECT * 
            FROM customer 
            WHERE store_id = 1'''
sql_to_df(query).head()
```

# Syntax for AND

```python
query = ''' SELECT *
            FROM film 
            WHERE release_year = 2006
            AND rating = 'R' '''
sql_to_df(query).head()
```

# Syntax for OR

```python
query = ''' SELECT * 
            FROM film 
            WHERE rating = 'PG'
            OR rating = 'R' '''
sql_to_df(query).head()
```

# Aggregate functions
AVG()

```python
query = '''SELECT COUNT(customer_id)
           FROM customer;'''
sql_to_df(query)
```

# SQL Wildcards
<br> three types:
<br> % (replace one or more characters)
<br> _ (one character)
<br> [charllist] (any characters in the [])

```python
query = ''' SELECT * 
            FROM customer 
            WHERE first_name LIKE 'M%'; '''
sql_to_df(query).head()
```

```python
query = ''' SELECT * 
            FROM customer 
            WHERE last_name LIKE '_ING'; '''
sql_to_df(query)
```

In MySQL you would use: 
<br> WHERE value LIKE '[charlist]%'
<br> In SQLite you use:
<br> WHERE value GLOB '[charlist]*'

```python
query = ''' SELECT * 
            FROM customer 
            WHERE first_name GLOB '[AB]*'; ''' # first_name either starts with A or B
sql_to_df(query).head()
```

# Order by
SELECT column_name 
<br> FROM table_name 
<br> ORDER BY column_name ASC|DESC

```python
query = ''' SELECT * 
            FROM customer 
            ORDER BY last_name;'''
sql_to_df(query).head()
```

```python
query = ''' SELECT * 
            FROM customer 
            ORDER BY last_name DESC;
'''
sql_to_df(query).head()
```

# SQL GROUP BY 
SELECT column_name, aggregate_function(column_name)
<br> FROM table_name
<br> WHERE column_name operator value 
<br> GROUP BY column_name

```python
query = ''' SELECT store_id, COUNT(customer_id)
            FROM customer 
            GROUP BY store_id;
'''
sql_to_df(query).head()
```
